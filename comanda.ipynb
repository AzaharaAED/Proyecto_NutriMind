```python
!pip install -q openai-whisper
import whisper

# 1. CONFIGURACIÓN

_WHISPER_MODEL = None
WHISPER_MODEL_NAME = "small"

# 2. CARGA DIFERIDA DE WHISPER

def get_whisper_model(model_name=WHISPER_MODEL_NAME):
    global _WHISPER_MODEL

    if _WHISPER_MODEL is None:
        _WHISPER_MODEL = whisper.load_model(model_name)

    return _WHISPER_MODEL


# 3. TRANSCRIPCIÓN DE AUDIO

def transcribir_audio(ruta_audio, language="es", model_name=WHISPER_MODEL_NAME):
    modelo_whisper = get_whisper_model(model_name=model_name)
    resultado = modelo_whisper.transcribe(ruta_audio, language=language)
    texto_detectado = resultado["text"].strip()

    return {
        "texto": texto_detectado,
        "idioma": language
    }


# 4. OPCIONES PRINCIPALES

def obtener_opciones_principales():
    return {
        "1": "ver_progreso",
        "2": "dieta_semanal",
        "3": "lista_compra",
        "4": "recomendar_receta",
        "5": "que_puedo_cocinar",
        "6": "otra_consulta"
    }


def obtener_menu_opciones():
    return {
        "1": "Ver mi progreso",
        "2": "Hacerme una dieta semanal",
        "3": "Lista de la compra para esta semana",
        "4": "Recomiéndame una receta",
        "5": "Qué puedo cocinar hoy",
        "6": "Otra consulta"
    }


def validar_opcion_principal(opcion):
    opciones = obtener_opciones_principales()
    return opciones.get(str(opcion).strip())


# 5. CONSTRUCCIÓN DE LA SALIDA

def construir_comanda(tipo_peticion=None, texto_usuario="", modo_entrada=None):
    return {
        "tipo_peticion": tipo_peticion,
        "texto_usuario": str(texto_usuario).strip(),
        "modo_entrada": modo_entrada
    }


# 6. FUNCIONES NO INTERACTIVAS

def obtener_comanda_para_modelo_desde_texto(tipo_peticion, texto_usuario=""):
    return construir_comanda(
        tipo_peticion=tipo_peticion,
        texto_usuario=texto_usuario,
        modo_entrada="texto" if str(texto_usuario).strip() else None
    )


def obtener_comanda_para_modelo_desde_audio(tipo_peticion, ruta_audio, model_name=WHISPER_MODEL_NAME):
    transcripcion = transcribir_audio(
        ruta_audio=ruta_audio,
        language="es",
        model_name=model_name
    )

    return construir_comanda(
        tipo_peticion=tipo_peticion,
        texto_usuario=transcripcion["texto"],
        modo_entrada="voz"
    )


def obtener_comanda_para_modelo_sin_especificacion(tipo_peticion):
    return construir_comanda(
        tipo_peticion=tipo_peticion,
        texto_usuario="",
        modo_entrada=None
    )


# 7. FUNCIONES INTERACTIVAS

def leer_texto_usuario_interactivo():
    texto = input("Escribe tu mensaje: ").strip()

    return {
        "texto": texto
    }


def seleccionar_opcion_principal_interactivo():
    opciones = obtener_opciones_principales()
    menu = obtener_menu_opciones()

    print("¿Qué quieres hacer hoy?")
    for clave, descripcion in menu.items():
        print(f"{clave}. {descripcion}")

    opcion = input("Selecciona una opción: ").strip()

    if opcion not in opciones:
        return None

    return opciones[opcion]


def elegir_modo_entrada_interactivo():
    print("\n¿Cómo quieres especificarlo?")
    print("1. Hablar")
    print("2. Escribir")

    opcion = input("Selecciona una opción: ").strip()

    if opcion == "1":
        return "voz"
    if opcion == "2":
        return "texto"

    return None


def quiere_especificar_mas_interactivo():
    respuesta = input("\n¿Quieres especificar más tu petición? (si/no): ").strip().lower()
    return respuesta in ["si", "sí", "s"]


def obtener_especificacion_adicional_interactiva(model_name=WHISPER_MODEL_NAME):
    modo = elegir_modo_entrada_interactivo()

    if modo == "voz":
        ruta_audio = input("Introduce el nombre o ruta del audio: ").strip()
        resultado = transcribir_audio(ruta_audio, language="es", model_name=model_name)

        return {
            "modo_entrada": "voz",
            "texto": resultado["texto"]
        }

    if modo == "texto":
        resultado = leer_texto_usuario_interactivo()

        return {
            "modo_entrada": "texto",
            "texto": resultado["texto"]
        }

    return {
        "modo_entrada": None,
        "texto": ""
    }


def obtener_comanda_usuario_interactivo(model_name=WHISPER_MODEL_NAME):
    opcion_principal = seleccionar_opcion_principal_interactivo()

    if opcion_principal is None:
        return construir_comanda(
            tipo_peticion=None,
            texto_usuario="",
            modo_entrada=None
        )

    texto_usuario = ""
    modo_entrada = None

    if quiere_especificar_mas_interactivo():
        especificacion = obtener_especificacion_adicional_interactiva(model_name=model_name)
        texto_usuario = especificacion["texto"]
        modo_entrada = especificacion["modo_entrada"]

    return construir_comanda(
        tipo_peticion=opcion_principal,
        texto_usuario=texto_usuario,
        modo_entrada=modo_entrada
    )


# 8. FUNCIÓN FINAL INTERACTIVA

def obtener_comanda_para_modelo_interactivo(model_name=WHISPER_MODEL_NAME):
    return obtener_comanda_usuario_interactivo(model_name=model_name)
```
